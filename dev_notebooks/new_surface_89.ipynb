```python
import torch
import torchlensmaker as tlm


mirror = tlm.SphereByCurvature(12, -0.05)

optics = tlm.Sequential(
    tlm.PointSourceAtInfinity(14),
    tlm.Gap(20),
    tlm.ReflectiveSurface(mirror, miss_mode="pass"),
    tlm.Gap(0),
)

optics.set_sampling2d(pupil=100)

controls={"show_optical_axis": True, "show_other_axes": True, "show_kinematic_joints": True}
scene = tlm.show2d(optics, end=12, controls=controls, return_scene=True)
```

```python
import torch
import torchlensmaker as tlm


# problem1: sharing surface submodule breaks rendering (because of by module indexing in output tree)

C = torch.nn.Parameter(torch.tensor(0.05))
mirror = tlm.SphereByCurvature(12, C)
mirror2 = tlm.SphereByCurvature(12, C, scale=-1)

optics = tlm.Sequential(
    tlm.PointSourceAtInfinity(14),
    tlm.Gap(20),
    tlm.RefractiveSurface(mirror, material="BK7", miss_mode="absorb"),
    tlm.Gap(2),
    tlm.RefractiveSurface(mirror2, material="air", miss_mode="absorb"),
    tlm.Gap(2),
    #tlm.RefractiveSurface(tlm.SphereByCurvature(12, C), material="air", miss_mode="absorb"),
    tlm.Gap(0),
)

optics.set_sampling2d(pupil=100)

controls={"show_optical_axis": True, "show_other_axes": True, "show_kinematic_joints": True}
scene = tlm.show2d(optics, end=18, controls=controls, return_scene=True)
```

```python
import torch
import torchlensmaker as tlm

# sphere by radius with negative radius and outer anchor 
mirror = tlm.SphereByRadius(12, -1/0.05)

optics = tlm.Sequential(
    tlm.PointSourceAtInfinity(10),
    tlm.Gap(10),
    tlm.RefractiveSurface(mirror.clone(anchors=(1, 0), scale=1), material="BK7", miss_mode="pass"),
)

optics.set_sampling2d(pupil=100)
optics.set_sampling3d(pupil=1000)

controls={"show_optical_axis": True, "show_other_axes": True, "show_kinematic_joints": True}
tlm.show2d(optics, end=12, controls=controls)
tlm.show3d(optics, end=12, controls=controls)
```

```python
import torch
import torchlensmaker as tlm


surface1 = tlm.SphereByCurvature(12, 0.05, num_iter=8, tol=1e-5)

lens = tlm.lenses.symmetric_singlet(surface1, tlm.OuterGap(0), material="BK7")

optics = tlm.Sequential(
    tlm.PointSourceAtInfinity(14),
    tlm.Gap(20),
    lens,
    tlm.Gap(0),
)
print(lens.inner_thickness())
print(lens.outer_thickness())

optics.set_sampling2d(pupil=100)

controls={"show_optical_axis": True, "show_other_axes": True, "show_kinematic_joints": True}
scene = tlm.show2d(optics, end=12, controls=controls, return_scene=True)
```
