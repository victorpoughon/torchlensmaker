```python
import torch
import torchlensmaker as tlm
import torch.optim as optim

# Setup two spherical surfaces with initial radiuses
surface1 = tlm.SphereByCurvature(diameter=30, C=1/-60)
surface2 = tlm.SphereByCurvature(diameter=30, C=1/-25)

lens = tlm.lenses.singlet(surface1, tlm.OuterGap(2.2), surface2, material="BK7")

focal = 120.5

# Build the optical sequence
optics = tlm.Sequential(
    tlm.ObjectAtInfinity(beam_diameter=8, angular_size=30),
    tlm.Gap(15),
    lens,
    tlm.Gap(focal),
    tlm.ImagePlane(diameter=100, magnification=125.),
)

tlm.show(optics, dim=2, end=10, reverse=False)
tlm.show(optics, dim=2, end=10, reverse=True)

# should reverse be a bool flag or a separate forward function?

# De we need a Reverse() element that swaps prograde and retrograde?
# careful: it will make a copy of its given module :)

# to compute front paraxial points:
# Sequence(
#    light_source   reversed
#    lens           reversed
#    <- output rays
# note that the full sequential (light + lens) is not reversed
```
