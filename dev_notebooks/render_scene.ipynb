```python
import torchlensmaker as tlm

m1 = tlm.NonDispersiveMaterial(1.517)
m2 = tlm.NonDispersiveMaterial(1.649)

doublet = tlm.Sequential(
    tlm.RefractiveSurface(tlm.SphereByCurvature(4.0, C=0.135327), materials=("air", m1)),
    tlm.Gap(1.05),
    tlm.RefractiveSurface(tlm.SphereByCurvature(4.0, C=-0.19311), materials=(m1, m2)),
    tlm.Gap(0.4),
    tlm.RefractiveSurface(tlm.SphereByCurvature(4.0, C=-0.06164), materials=(m2, "air")),
)


trace = tlm.ModelTrace.empty(dim=2)
outputs = doublet.forward_trace(tlm.SequentialData.empty(dim=2), "", trace)

scene = tlm.render_model_trace(trace)

tlm.display_scene(scene)
```
