```python
import torchlensmaker as tlm
import torch

lens1 = tlm.lenses.symmetric_singlet(
    tlm.Conic(10.0, C=-0.05, K=0.95), tlm.InnerGap(1.0), material="BK7", anchors=(1, 1)
)
lens2 = tlm.lenses.symmetric_singlet(
    tlm.Parabola(8.0, A=-0.03), tlm.InnerGap(1.0), material="BK7", anchors=(1, 1)
)

tlm.show2d(
    tlm.Sequential(
        tlm.PointSourceAtInfinity(4), tlm.Gap(4), lens1, tlm.Gap(0.0), lens2
    ),
    end=4,
)
```
