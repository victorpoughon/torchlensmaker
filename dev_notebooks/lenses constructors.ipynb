```python
import torchlensmaker as tlm
import torch

lens1 = tlm.lenses.symmetric_singlet(tlm.Conic(10.0, C=-0.05, K=0.95), tlm.InnerGap(1.0), material="water", anchors=(0, 1))
lens2 = tlm.lenses.singlet(
    tlm.SphereByRadius(8.0, R=-10.0),
    tlm.InnerGap(0.5),
    tlm.Parabola(8.0, A=0.01, normalize=False),
    material="BK7",
    anchors=(1, 0),
)

tlm.show2d(tlm.Sequential(lens1, tlm.Gap(0.5), lens2))
```
