```python
import json
import numpy as np
import matplotlib as mpl
import colorcet as cc

#cmap = mpl.colormaps['rainbow']
cmap = cc.cm.CET_I2

def cmap255(x):
    col = cmap(x)
    return list(map(int, [c*255 for c in col[:3]]))

points = []

start = [0, 0]

R = 30
N = 50
theta = np.linspace(-np.pi/3, np.pi/3, N)

for i, t in enumerate(theta):
    k = i/N
    
    end = [R*np.cos(t), R*np.sin(t)]
    points.append([
        *start,
        *end,
        *cmap255(i/N)
    ])

json.dumps(points)
```

## colormap to css gradient

```python
# css gradient from mpl color map

# option1
#cmap = mpl.colormaps['plasma']
#start, end = 0.5, 0.90

# option2
cmap = mpl.colormaps['rainbow']
start, end = 0.0, 1.0


N = 8

def pl(v):
    return map(int, np.round(np.array(cmap(v))*255))

space = np.linspace(start, end, N)
domain = np.linspace(0, 1, N)

s = "background: linear-gradient(-90deg, "
for i, j in zip(space, domain):
    r, g, b, a = pl(i)
    s += f"rgba({r},{g},{b},1) {round(100*j)}%, "

s = s[:-2]
s += ");"
print(s)
```
