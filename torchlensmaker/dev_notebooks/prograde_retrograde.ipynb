```python
import torch
import torchlensmaker as tlm
import torch.optim as optim

# Setup two spherical surfaces with initial radiuses
surface1 = tlm.SphereByCurvature(diameter=30, C=1/-60)
surface2 = tlm.SphereByCurvature(diameter=30, C=1/-25)

# Build the optical sequence
lens = tlm.Sequential(
    tlm.RefractiveSurface(tlm.KinematicSurface(surface1, anchors=(0, 1)), materials=("air", "BK7")),
    tlm.Gap(1),
    tlm.RefractiveSurface(tlm.KinematicSurface(surface2, anchors=(1, 0)), materials=("BK7", "air")),
)

print(lens)

controls = {"show_optical_axis": True, "show_other_axes": True}
tlm.show(lens, dim=2, end=10, controls=controls, reverse=False)
tlm.show(lens, dim=2, end=10, controls=controls, reverse=True)
#tlm.show(optics, dim=3, end=10)
```

```python
import torch
import torchlensmaker as tlm
import torch.optim as optim

# Setup two spherical surfaces with initial radiuses
surface1 = tlm.SphereByCurvature(diameter=30, C=1/-60)
surface2 = tlm.SphereByCurvature(diameter=30, C=1/-25)

lens = tlm.lenses.singlet(surface1, tlm.OuterGap(2.2), surface2, material="BK7")

focal = 120.5

# Build the optical sequence
optics = tlm.Sequential(
    tlm.ObjectAtInfinity(beam_diameter=8, angular_size=30),
    tlm.Gap(15),
    lens,
    tlm.Gap(focal),
    tlm.ImagePlane(diameter=100, magnification=125.),
)

print(optics)

controls = {"show_optical_axis": True, "show_other_axes": True}
tlm.show(lens, dim=2, end=10, controls=controls, reverse=False)
tlm.show(lens, dim=2, end=10, controls=controls, reverse=True)
#tlm.show(optics, dim=3, end=10)
```
