```python
import torchlensmaker as tlm
import torch


optics = tlm.Sequential(
    tlm.PointSource(90),
    tlm.Gap(1.5),
    tlm.RefractiveSurface(tlm.SphereByRadius(3, R=2), materials=("water", "air"), tir_mode="reflect"),
    tlm.Gap(0),
)

optics.set_sampling2d(pupil=100)

tlm.show2d(optics, end=5)

#tlm.RefractiveSurface(tlm.Sphere(
```
